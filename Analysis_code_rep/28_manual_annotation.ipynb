```python
mdata['rna'].obs['celltype_v2'].cat.categories
```

```python
healthy_ct=['BMCP', 'Erythroblast', 'GMP', 'HSC', 'MEP','MEP/MKP', 'MPP', 'Monocyte progenitor']
atypical_ct=['Atypical cluster O', 'Atypical cluster a',
       'Atypical cluster b', 'Atypical cluster c', 'Atypical cluster d',
       'Atypical cluster e', 'Atypical cluster f', 'Atypical cluster g',
       'Atypical cluster h', 'Atypical cluster i', 'Atypical cluster j',
       'Atypical cluster k', 'Atypical cluster l', 'Atypical cluster m',
       'Atypical cluster n']
mdata['rna'].obs['celltype_category'] = mdata['rna'].obs['celltype_v2'].copy()
```

```python
mdata['rna'].obs['celltype_category'] = mdata['rna'].obs['celltype_category'].cat.add_categories(['Healthy', 'Atypical'])
mdata['rna'].obs['celltype_category'].loc[mdata['rna'].obs['celltype_category'].isin(healthy_ct)] = 'Healthy'
mdata['rna'].obs['celltype_category'].loc[mdata['rna'].obs['celltype_category'].isin(atypical_ct)] = 'Atypical'
mdata['rna'].obs['celltype_category'] = mdata['rna'].obs['celltype_category'].cat.remove_unused_categories()
mdata['rna'].obs['celltype_category'].value_counts()
```
