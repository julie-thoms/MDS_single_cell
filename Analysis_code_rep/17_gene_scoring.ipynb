```python
import numpy as np

p18 = df[df['patient_alias'] == "P18"]
# Ensure correct order for timepoints
timepoint_order = ["pre", "mid", "progression"]
p18["timepoint_coarse"] = pd.Categorical(p18["timepoint_coarse"], categories=timepoint_order, ordered=True)

# Count cells per (celltype, timepoint, phase)
counts = p18.groupby(["infercnv_new_celltype", "timepoint_coarse", "phase"]).size().reset_index(name="count")

# Normalize within each (celltype, timepoint) group
totals = counts.groupby(["infercnv_new_celltype", "timepoint_coarse"])["count"].transform("sum")
counts["proportion"] = counts["count"] / totals

# Pivot to prep for stacked plotting
pivot_df = counts.pivot_table(
    index=["infercnv_new_celltype", "timepoint_coarse"],
    columns="phase",
    values="proportion",
    fill_value=0
).reset_index()

# Ensure consistent phase order
phase_order = ['G1', 'S', 'G2M']
pivot_df = pivot_df[["infercnv_new_celltype", "timepoint_coarse"] + phase_order]

# Plot
celltypes = pivot_df["infercnv_new_celltype"].unique()
n_cols = 4
n_rows = -(-len(celltypes) // n_cols)  # ceiling division

fig, axes = plt.subplots(n_rows, n_cols, figsize=(4 * n_cols, 4 * n_rows), sharey=True)
axes = axes.flatten()

colors = sns.color_palette("pastel", n_colors=len(phase_order))

for i, celltype in enumerate(celltypes):
    ax = axes[i]
    sub_df = pivot_df[pivot_df["infercnv_new_celltype"] == celltype]
    bottoms = np.zeros(len(sub_df))
    
    for j, phase in enumerate(phase_order):
        ax.bar(
            sub_df["timepoint_coarse"],
            sub_df[phase],
            bottom=bottoms,
            color=colors[j],
            label=phase if i == 0 else ""  # only label once
        )
        bottoms += sub_df[phase].values

    ax.set_title(celltype)
    ax.set_ylim(0, 1)
    ax.set_xlabel("Timepoint")
    if i % n_cols == 0:
        ax.set_ylabel("Proportion")

# Clean up
for ax in axes[len(celltypes):]:
    ax.axis("off")

fig.legend(phase_order, title="Cell Cycle Phase", loc="upper right", bbox_to_anchor=(1.05, 0.95))
fig.suptitle("100% Stacked Barplot of Cell Cycle Phase per Timepoint (per Cell Type) P18", y=1.02)
plt.tight_layout()
plt.show()
```
