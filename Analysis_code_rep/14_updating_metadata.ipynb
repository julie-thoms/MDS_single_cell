# Updating Metadata

```python
import scanpy as sc
import pandas as pd
import muon as mu
import numpy as np
import anndata as ad
import matplotlib.pyplot as plt
import seaborn as sns
import warnings


warnings.filterwarnings('ignore')
sc.settings.verbosity = 3
plt.rcParams['figure.dpi'] = 200
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['figure.figsize'] = (5, 5)
```

```python
adata= sc.read_h5ad("HSC_gex_MDSpaper_figure2_newcelltypes_updatedtime.h5ad")
adata
```

```python
adata.obs[['patient','patient_alias']].value_counts()
```

```python
#Import csv file
patient_csv = pd.read_csv("data/patient_alias_updateMar2025.csv")
patient_csv
```

```python
#create mapping dictionary
patient_dict = dict(zip(patient_csv['PID'].astype(str), patient_csv['alias']))
patient_dict
```

```python
adata.obs['patient_alias'] = adata.obs['patient'].map(patient_dict, na_action='ignore')
```

```python
adata.obs[['patient','patient_alias']].value_counts()
```

```python
adata.obs['new_celltype'].value_counts()
```

```python
#change cluster 17 to BMCP

# Convert 'new_celltype' to categorical if it's not already
if isinstance(adata.obs['new_celltype'].dtype, pd.CategoricalDtype):
    # Add new category before setting values
    adata.obs['new_celltype'] = adata.obs['new_celltype'].cat.add_categories(["BMCP"])

#set value in new_celltype column
adata.obs.loc[adata.obs['leiden'] == "17", 'new_celltype'] = "BMCP"
adata.obs[['new_celltype','leiden']].value_counts()
```

```python
sc.pl.umap(adata, color=['leiden','new_celltype'], legend_loc="on data")
```
